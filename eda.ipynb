```python
import pandas as pd
```

```python
# Read CSV normally without parsing dates yet
ohlc_file = "data/ohlc_adjusted_17_fixed.csv"
df = pd.read_csv(ohlc_file, header=[0])  # Use first row as column headers

# Drop the second row (which contains NaNs)
df = df.iloc[1:].reset_index(drop=True)

# Rename the first column to 'date'
df.rename(columns={"Unnamed: 0": "date"}, inplace=True)

# Convert 'date' to datetime
df["date"] = pd.to_datetime(df["date"], errors="coerce")

# Set 'date' as index
df.set_index("date", inplace=True)

# Convert numerical columns to float
df = df.apply(pd.to_numeric, errors="coerce")

# Display cleaned dataset info
print(df.info())
print(df.columns)
print(df.head())
```

```python
# Define the new suffixes
suffixes = ['open', 'high', 'low', 'close']

# Rename columns
new_columns = {}
for col in df.columns:
    base_name = col.split('.')[0]  # Extract the base name (before the first dot)
    index = int(col.split('.')[1]) if '.' in col else 0  # Extract the index after the dot
    new_columns[col] = f"{base_name}.{suffixes[index]}"

df.rename(columns=new_columns, inplace=True)

print(df.columns)
```
